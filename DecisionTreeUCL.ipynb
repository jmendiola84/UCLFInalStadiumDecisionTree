# Decision Tree Classifier - UEFA Champions League
----------------
Este codigo realiza una clasificacion de arbol de decision para un set de datos que incluye informacion sobre estadios de la UEFA con 3 variables de decision y una variable objetivo para definir las probabilidades de ser sede de la final de la UEFA Champions League.

## Importar Librerias

```python
from sklearn.cross_validation import train_test_split 
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from lightgbm import create_tree_digraph
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
```

## Leer archivo CSV y crear dataframe

```python
stadiums_csv = "UEFA_Stadiums_Geo_Data_UTF8.csv"
stadiums_df = pd.read_csv(stadiums_csv)
stadiums_df.dtypes
```

## Remover columnas que no se usaran en el modelo

```python
stadiums_df = stadiums_df.drop(['Unnamed: 0','Confederation','City','HomeTeams','IOC','Capacidad/1000Habs','Latitude', 'Longitude'], axis=1)

stadiums_df.head()
```

## Asignar los valores de las variables de decision y la variable objetivo

```python
X = stadiums_df.values[:, [1,3,4]]
Y = stadiums_df.values[:, 6]
Y=Y.astype('int')
```

## Dividir el data set en Training y Test

```python
X_train, X_test, y_train, y_test = train_test_split(  
    X, Y, test_size = 0.3, random_state = 100) 
```

## Generar el arbol de decision en base a los datos proporcionados

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

## Visualizacion del arbol de decision

```python
import graphviz 
dot_data = tree.export_graphviz(
    clf, out_file=None, 
    feature_names=['Stadium Size','Country Population','UCLChampions'],  
    class_names=['No Candidate','Candidate'],  
    filled=True, rounded=True,  
    special_characters=True)  

import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png('UCLDecisionTree.png')

graph = graphviz.Source(dot_data)  
graph 
```
